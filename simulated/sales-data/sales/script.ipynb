```python
import numpy as np
import pandas as pd
```

```python

def formatRandId(
        id, 
        numericPrefix = 0, 
        numberOfDigitis = 4,
        prefix = "",
        suffix = ""
):
    return prefix + str(int(np.round((id + numericPrefix) * (10 ** (numberOfDigitis - 1))))) + suffix
```

```python
class Sale:
    def __init__(
            self,
            date,
            clientId,
            clientName,
            clientType,
            productId,
            productName,
            productCategory,
            productPrice,
            saleType,
            saleAmount,
            saleDiscount,
    ):
        self.date = date
        self.clientId = clientId
        self.clientName = clientName
        self.clientType = clientType
        self.productId = productId
        self.productName = productName
        self.productCategory = productCategory
        self.productPrice = productPrice
        self.saleType = saleType
        self.saleAmount = saleAmount
        self.saleDiscount = saleDiscount

    def to_dict(self):
        return {
            "date": self.date,
            "clientId": self.clientId,
            "clientName": self.clientName,
            "clientType": self.clientType,
            "productId": self.productId,
            "productName": self.productName,
            "productCategory": self.productCategory,
            "productPrice": self.productPrice,
            "saleType": self.saleType,
            "saleAmount": self.saleAmount,
            "saleDiscount": self.saleDiscount,
        }
```

```python
def ub(x):

    x_len = x.shape[0]

    ub_all = np.zeros(x_len)
    ub_curr = 0

    for i in np.arange(x_len):
        ub_curr += float(x[i][1])
        ub_all[i] = ub_curr

    return ub_all
```

```python
def to_dscrt(x, v):
    b = ub(v)

    for i in np.arange(len(b)):
        if x < b[i]:
            return v[i][0]

def get_to_dscrt(v):

    def f(x):
        return to_dscrt(x, v)

    return f
```

```python
numOfSales = 10000
bins = 10

sales_types_frq_tbl = np.array([
    ["Online sales", .44],
    ["Direct sales", .18],
    ["Partner sales", .19],
    ["Referral sales", .09],
    ["Subscription sales", .1],
])

client_types_frq_tbl = np.array([
    ["Individuals", .46],
    ["Businesses", .21],
    ["Non-profit", .05],
    ["Mixed", .28],
])

to_dscrt_sales_types = np.frompyfunc(get_to_dscrt(sales_types_frq_tbl), 1, 1)
to_dscrt_client_types = np.frompyfunc(get_to_dscrt(client_types_frq_tbl), 1, 1)

rg = np.random.default_rng()
```

```python
numOfSales = 1000
transactionId = np.random.uniform(0, 1, numOfSales)
productId = np.random.normal(.5, .05, numOfSales)
clientId = np.random.uniform(0, 1, numOfSales)
listedPrice = np.random.normal(14000, 4000, numOfSales)
saleAmount = np.random.normal(13500, 5000, numOfSales)
sales_types = to_dscrt_sales_types(rg.random((numOfSales)))
client_types = to_dscrt_client_types(rg.random((numOfSales)))

sales = []

for i in np.arange(numOfSales):
    sales.append(Sale(
        date = "2024-01-01",
        clientId = formatRandId(clientId[i], 0, 4), #, "AP"),
        clientName = "John Doe",
        clientType = client_types[i],
        productId = formatRandId(productId[i], 0, 4),
        productName = "Widget Deluxe",
        productCategory = "Gadgets",
        productPrice = listedPrice[i],
        saleType= sales_types[i],
        saleAmount = saleAmount[i],
        saleDiscount = "10%",

    ))

df_sales = pd.DataFrame([x.to_dict() for x in sales])

df_sales.to_csv("sales-data-sales.csv")

df_sales
```
