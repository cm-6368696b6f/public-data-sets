```python
import numpy as np
import pandas as pd
```

```python
class Sale:
    def __init__(
            self,
            date,
            clientName,
            clientType,
            productId,
            productName,
            productCategory,
            productPrice,
            saleAmount,
            saleDiscount,
    ):
        self.date = date
        self.clientName = clientName
        self.clientType = clientType
        self.productId = productId
        self.productName = productName
        self.productCategory = productCategory
        self.productPrice = productPrice
        self.saleAmount = saleAmount
        self.saleDiscount = saleDiscount

    def to_dict(self):
        return {
            "date": self.date,
            "clientName": self.clientName,
            "clientType": self.clientType,
            "productId": self.productId,
            "productName": self.productName,
            "productCategory": self.productCategory,
            "productPrice": self.productPrice,
            "saleAmount": self.saleAmount,
            "saleDiscount": self.saleDiscount,
        }
```

```python
sales = []

for i in np.arange(100):
    sales.append(Sale(
        date = "2024-01-01",
        clientName = "John Doe",
        clientType = "Premium",
        productId = "12345",
        productName = "Widget Deluxe",
        productCategory = "Gadgets",
        productPrice = "$199.99",
        saleAmount = "50 units",
        saleDiscount = "10%",

    ))

df_sales = pd.DataFrame([x.to_dict() for x in sales])
df_sales
```

```python
df_sales.to_csv("sales-data-sales.csv")
```
